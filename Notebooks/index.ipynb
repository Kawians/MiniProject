# Let's play Rock Paper Scissors against Neural Network

```python
from PIL import Image
from matplotlib import pyplot as plt

img = Image.open('../images/rock_paper_scissors_cnn.jpg')
fig = plt.imshow(img)

plt.axis('off')
fig.axes.get_xaxis().set_visible(False)
fig.axes.get_yaxis().set_visible(False)

plt.tight_layout()
plt.show()
```
